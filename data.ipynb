```python
result_path = './data/result.txt'
# load in the result
with open(result_path, 'r') as f:
    result = f.read()
print(result)
type(result)
```

```python
lines = result.split('\n')
df = pd.DataFrame([line.split() for line in lines[1:]], columns=lines[0].split())
df.head()
```

```python
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Convert columns to appropriate data types
df['gen'] = df['gen'].astype(int)
df['avg'] = df['avg'].astype(float)
df['min'] = df['min'].astype(float)
df['max'] = df['max'].astype(float)
df['std'] = df['std'].astype(float)

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='gen', y='avg', data=df, label='avg', marker='o')
sns.lineplot(x='gen', y='min', data=df, label='min', marker='o')
sns.lineplot(x='gen', y='max', data=df, label='max', marker='o')
sns.lineplot(x='gen', y='std', data=df, label='std', marker='o')

# Customize the plot
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.legend()
plt.title('Fitness over Generation')
plt.show()
# Create area plots
plt.figure(figsize=(10, 6))
df.set_index('gen')[['avg', 'min', 'max', 'std']].plot(kind='area', stacked=False)
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.legend()
plt.title('Fitness over Generation (Area Plot)')
plt.show()
```

```python
# Create the plot with log scale
plt.figure(figsize=(10, 6))
sns.lineplot(x='gen', y='avg', data=df, label='avg', marker='o')
sns.lineplot(x='gen', y='min', data=df, label='min', marker='o')
sns.lineplot(x='gen', y='max', data=df, label='max', marker='o')
sns.lineplot(x='gen', y='std', data=df, label='std', marker='o')

# Customize the plot
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.yscale('log')
plt.legend()
plt.title('Fitness over Generation (Log Scale)')
plt.show()

# Create area plots with log scale
plt.figure(figsize=(10, 6))
df.set_index('gen')[['avg', 'min', 'max', 'std']].plot(kind='area', stacked=False)
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.yscale('log')
plt.legend()
plt.title('Fitness over Generation (Area Plot, Log Scale)')
plt.show()
```
